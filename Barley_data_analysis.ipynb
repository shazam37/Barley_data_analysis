# Data Preprocessing and Transformation

We begin by dividing our dataset into dependent (y) and independent variables (X)

y = barley variety column

X = all the wavelength columns

```python
X = df.drop(['Predictor'],axis=1)
y = df['Predictor']
```

```python
y
```

Since we have categorical values in y. We need to apply Labelencoder to encode them before applying any ML algorithm

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

We ensure which variety is encoded to what numerical value

```python
for category, encoded_value in zip(y, y_encoded):
    print(f"{category} -> {encoded_value}")
```

Now we move onto splitting and scaling our values. We first split the dataset with a test-size of 20% and then Standardize
them using StandardScaler

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=101)
```

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

# Model Selection

We will check multiple ML algorithms to see which one performs the best. The algorithms we will be testing are:
1. Logistic Regression
2. K-Nearest Neighbour (KNN)
3. Support Vector Machine (SVM)
4. Naive Bayes
5. Random Forest Classifier
6. AdaBoost Classifier
7. Artificial Neural Network (ANN)

We will cross-validate where necessary

We first begin with Logistic Regression:

```python
from sklearn.linear_model import LogisticRegressionCV

'''Given the small size of our dataset, we will cross-validate 5 times, use l2 as penalty, use lbfgs solver,
with a max iteration of 1000 '''

lc = LogisticRegressionCV(cv=5, penalty='l2', solver='lbfgs', max_iter=1000)
lc.fit(scaled_X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix, accuracy_score
```

```python
pred = lc.predict(scaled_X_test)
accuracy_score(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

Next we move on to K-Nearest Neighbour. We begin by checking out the optimum number of neghbours for our problem.
We plot out the Error-rate v/s K-value curve. We check where we are getting the least error

```python
from sklearn.neighbors import KNeighborsClassifier

test_error_rates = []

for k in range(1,100):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_X_train,y_train)

    y_pred_test = knn_model.predict(scaled_X_test)

    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)

plt.figure(figsize=(10,6),dpi=200)
plt.plot(range(1,100),test_error_rates,label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")
```

It seems that we are getting k=1 as the optimum neghbour value having least error. Lets check further

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(scaled_X_train,y_train)

pred = knn.predict(scaled_X_test)
accuracy_score(y_test,pred)
```

We are getting a horrible accuracy on KNN. Thus KNN is not suitable for our given problem

We next move on to SVM. We will do a grid search cross validation to check for the most optimum values:

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

''' We will go with different values of C and will check for both linear as well as rbf kernels with different values
for gamma'''

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': [0.1, 1, 10],
}

svm = SVC()
grid = GridSearchCV(svm, param_grid, cv=5)
grid.fit(scaled_X_train,y_train)
```

```python
grid.best_score_
```

```python
  grid.best_params_
```

```python
'''Through Grid Search, we found out the values of C, gamma, and kernel. We take the values of gamma and kernel as it is, but
we take the C value to 10000'''

svc = SVC(C=1000, kernel='linear',gamma=0.1)
svc.fit(scaled_X_train,y_train)
pred = svc.predict(scaled_X_test)
accuracy_score(y_test,pred)
```

Lets now move on to Naive Bayes Classifier. Since there arent any hyperparameters to tune (except for alpha values), we
fit directly:

```python
from sklearn.naive_bayes import GaussianNB

nb_classifier = GaussianNB()
nb_classifier.fit(scaled_X_train,y_train)
pred = nb_classifier.predict(scaled_X_test)
accuracy_score(y_test,pred)
```

With a 5% accuracy, Naive Bayes is not suitable for our problem

We move on to RandomForest Classifier, a bagging model which is expected to give us better result. We again validate
through GridSearchCV here:

```python
from sklearn.ensemble import RandomForestClassifier

n_estimators=[64,100,128,200]
max_features= [2,3,4]
bootstrap = [True,False]
oob_score = [True,False]

param_grid = {'n_estimators':n_estimators,
             'max_features':max_features,
             'bootstrap':bootstrap,
             'oob_score':oob_score}

rfc = RandomForestClassifier()
grid = GridSearchCV(rfc,param_grid)
grid.fit(scaled_X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
'''Since we are getting a very low accuracy on GridSearchCV result, we test without any hyperparameters'''

rfc = RandomForestClassifier()
rfc.fit(scaled_X_train,y_train)
pred = rfc.predict(scaled_X_test)
accuracy_score(y_test,pred)
```

RandomForest gives us a very low accuracy. Its also not suitable for our problem

We next move on to a boosting model (AdaBoost). Lets see how it performs

```python
from sklearn.ensemble import AdaBoostClassifier

abc = AdaBoostClassifier()
abc.fit(scaled_X_train,y_train)
pred = abc.predict(scaled_X_test)
accuracy_score(y_test,pred)
```

Again with a horrible accuracy of 10%, even boosting model is not fit for our case

We finally test Artificial Neural Network. We will test with 3 hidden layers having rectifier activation function
for the input layers with l2 regularization imposed. For final layer we go with softmax activation function owing to
large number of categorical features in our y_train:

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers

model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(scaled_X_train.shape[1],)))
model.add(Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(24, activation='softmax'))

# We Compile the model with adam optimizer and sparse_categorical_crossentropy loss function
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# We Train the model with early stopping based on validation loss
model.fit(scaled_X_train, y_train, epochs=1000, batch_size=32, verbose=1)

# We evaluate the model on the test set
loss, accuracy = model.evaluate(scaled_X_test, y_test, verbose=0)
print('Test Loss:', loss)
print('Test Accuracy:', accuracy)
```

We are getting a low accuracy on ANN. It might be because we have limited size of training data.
ANN works well for big and complex datasets

After testing for 7 different alogrithms, we finally get the most appropriate accuracy in the case of two algorithms:

1. Logistic Regression = 82%
2. Support Vector Classifier = 84%
